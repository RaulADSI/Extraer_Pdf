```python
import pdfplumber
import pandas as pd
import glob
import os

# Inicializar listas
names = []
tables = []

ext = "*.pdf"
PATH = "/content/drive/MyDrive/Prueba PDF"

# Obtener todos los archivos PDF en el directorio
all_files = []
for dirpath, dirnames, filenames in os.walk(PATH):
    all_files += glob.glob(os.path.join(dirpath, ext))

for idx, filename in enumerate(all_files):
    try:
        # Leer el PDF
        with pdfplumber.open(filename) as pdf:
            # Trabajar con la primera página
            first_page = pdf.pages[0]
            # Extraer tablas de la primera página
            page_tables = first_page.extract_tables()

            # Agregar el nombre del archivo a la lista
            names.append(filename)

            # Agregar tablas extraídas a la lista
            tables.append(page_tables)

    except Exception as e:
        print(f"Error procesando el archivo {filename}: {e}")

# Convertir las tablas extraídas en DataFrames y combinarlas en uno solo
df_list = []
for table in tables:
    for sub_table in table:
        df_list.append(pd.DataFrame(sub_table))

df = pd.concat(df_list, ignore_index=True)
df

```
